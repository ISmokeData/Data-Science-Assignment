```python
'''
Q1. load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.
Code-
'''
import seaborn as sns
import plotly.graph_objects as go
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = titanic['age'], y = titanic['fare'], mode= 'markers'))
fig.show()
```

```python
'''
Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
Code-
'''
tips = sns.load_dataset('tips')
```

```python
fig = go.box(titanic, y = 'tip')
fig.show()
```

```python
tips
```

```python
plot = go.Figure()
plot.add_trace(go.Box(y = tips['tip']))
plot.show()
```

```python
'''
Q3. Using the tips dataset in the plotly library, plot a histogram for x = "sex"  and y = "total_bill" column in the tips dataset. Also use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.
'''
import plotly.express as px
```

```python
df = px.data.tips()
fig = px.histogram(df, x = "sex" , y = "total_bill",pattern_shape= "smoker", color = "day")
fig.show()
```

```python
'''
Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
the color parameter.
Code-
'''
df = px.data.iris()
fig = px.scatter(df, x = "sepal_width", y = "sepal_length", color= "species", size= "petal_length",hover_data=['petal_width'])
fig.show()
```

```python
'''
Q5. What is Displot? Using Plotly express, plot a displot.
Ans - Several representation of statistical distributions are available in plotly. Such as histogram, violin plots, box plots etc. It is also possibke to combine several representations in the same plot.
Code-
'''
df = px.data.tips()
fig =  px.histogram(df, x="total_bill", y="tip", color="sex", marginal="rug",hover_data=df.columns)
fig.show()
```
